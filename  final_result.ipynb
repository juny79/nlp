```python
import matplotlib.pyplot as plt
import numpy as np

# Solar_v3  
v3_models = ['solar_qlora', 'solar_tuned', 'solar_gridsearch', 'solar_v2', 'solar_v3',
             'solar_v3_postprocessed', 'solar_v3_microtuned', 'solar_v3_ultrafine']
v3_middle_scores = [44.5387, 51.6863, 50.7397, 48.5028, 51.8026, 51.9393, 51.9421, 51.9302]
v3_last_scores   = [42.1096, 50.6476, 49.8780, 47.3276, 50.5183, 50.6190, 50.6215, 50.6245]

# Solar_v4  
v4_models = ['solar_v4', 'solar_v4_no_hallucination']
v4_middle_scores = [51.7703, 51.7905]
v4_last_scores   = [51.4181, 51.4457]

plt.figure(figsize=(14, 6))

# x 
x_v3 = np.arange(len(v3_models))
x_v4 = np.arange(len(v3_models), len(v3_models) + len(v4_models))

# v3: middle (dashed) / last (solid)
plt.plot(x_v3, v3_middle_scores, marker='o', linewidth=2, markersize=8, linestyle='--', label='v3 middle (dashed)', color='#2E86AB')
plt.plot(x_v3, v3_last_scores,   marker='^', linewidth=2, markersize=8, linestyle='-',  label='v3 last (solid)',    color='#2E86AB', alpha=0.9)

# v4: middle (dashed) / last (solid)
plt.plot(x_v4, v4_middle_scores, marker='s', linewidth=2, markersize=8, linestyle='--', label='v4 middle (dashed)', color='#A23B72')
plt.plot(x_v4, v4_last_scores,   marker='d', linewidth=2, markersize=8, linestyle='-',  label='v4 last (solid)',    color='#A23B72', alpha=0.9)

# xticks:   ( )
all_models = v3_models + v4_models
plt.xticks(np.arange(len(all_models)), all_models, rotation=45, ha='right')

plt.ylabel('Final Result ()', fontsize=12)
plt.xlabel(' ', fontsize=12)
plt.title('  Final Result  ', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=10)

#  (  middle/last   )   
per_model_max = [max(a,b) for a,b in zip(v3_middle_scores, v3_last_scores)] + [max(a,b) for a,b in zip(v4_middle_scores, v4_last_scores)]
max_score = max(per_model_max)
max_idx = per_model_max.index(max_score)

#  middle / last   (/ )
all_middle = v3_middle_scores + v4_middle_scores
all_last   = v3_last_scores + v4_last_scores
middle_max = max(all_middle)
middle_max_idx = all_middle.index(middle_max)
middle_max_model = all_models[middle_max_idx]
last_max   = max(all_last)
last_max_idx = all_last.index(last_max)
last_max_model = all_models[last_max_idx]

# /     
plt.scatter(middle_max_idx, middle_max, color='green', s=160, zorder=6, marker='*', label=f'middle : {middle_max:.4f} ({middle_max_model})')
plt.scatter(last_max_idx, last_max,     color='orange', s=140, zorder=6, marker='X', label=f'last : {last_max:.4f} ({last_max_model})')
plt.annotate(f'{middle_max:.4f}', (middle_max_idx, middle_max), textcoords="offset points", xytext=(0,10), ha='center', color='green')
plt.annotate(f'{last_max:.4f}', (last_max_idx, last_max),     textcoords="offset points", xytext=(0,10), ha='center', color='orange')

#  per-model  ()
plt.axhline(y=max_score, color='red', linestyle='--', alpha=0.3,
            label=f' : {max_score:.4f}')

plt.scatter(max_idx, max_score, color='red', s=200, zorder=5, alpha=0.5)

plt.tight_layout()
plt.show()

# :   
print(f"\nlast  : {last_max:.4f} ({last_max_model})")
print(f"v3 last : {np.mean(v3_last_scores):.4f}")
print(f"v4 last : {np.mean(v4_last_scores):.4f}")
print(f"\nmiddle  : {middle_max:.4f} ({middle_max_model})")
print(f" (/  ): {max_score:.4f} ({all_models[max_idx]})")
```
